<h4> Background </h4> <br></br>
The shortest path length is the idea that for any network of nodes connected by edges, there is a particular series of steps that can be taken to get from a starting node to an ending node in the shortest amount of time or in the fewest number of steps across edges [4]. <br></br><br></br>
In time-independent networks, the magnitudes of the paths between nodes are arbitrary. Finding the shortest path length in a network like this can be useful for analyzing social networks, where the paths between nodes represent not distances, but rather relationships. Analysis of this kind of network has proven useful in areas like anthropology, biology, economics, and communications [5]. <br></br><br></br>
Alternatively, in time-dependent networks, finding the shortest path length between nodes means finding the series of steps between them that covers the minimum total distance. This would be particularly useful for modeling systems like road networks. For example, the handling of city emergencies depends heavily on finding the quickest routes between destinations. GPS systems also function by mapping out the quickest routes for drivers depending on where they are at any given moment [1, 2]. Some other real-world applications of the shortest path problem would include order-picking in warehouses, computer wiring, drilling of circuit boards, x-ray crystallography, and vehicle routing for postal workers, school bus drivers, etc. [3].

Instead, I scaled the campus map so that it most closely represented our measurements as possible. Further research online suggested that this was a commonly-used and satisfactory alternative. This allowed me to then overlay our network onto centimeter graphing paper where each unit corresponded to one minute of travel time, which was very important for the next piece of code. <br></br><br></br>
For this step I used pieces of code that would find solutions for the shortest path through each node based on coordinates. The newly-adjusted network overlaid on the graph paper allowed me to assign each building a coordinate on the Cartesian plane with the maintenance building as the origin. I plugged these into our program that was improved upon by Dr. Cook to find two unique solutions to the problem. First is the Exhaustive Traveling Salesman solution which records the time it takes to travel the network in each of the possible different ways by brute force and then returns the shortest of them all. This code takes roughly 45 minutes to execute on CoCalc but avoided the memory issues we were encountering before modifying it as suggested. Second is the Greedy Traveling Salesman solution, a simpler and quicker alternative. It starts at the origin and then travels to the next node closest to that point. From there it travels to the closest unvisited node from that new point, and so on, until all the nodes have been visited. This algorithm has been shown to take only 25 percent more time on average than the true traveling salesman solution. <br></br><br></br>
It should also be noted that the user of this code may simply remove any coordinates they choose if they only need to visit a subset of the buildings I included in this project.

<br></br>

```python
import doctest
from itertools import permutations
import numpy as np
from scipy.spatial.distance import pdist, squareform
from pprint import pprint


def main():
    doctest.testmod()
    points = np.array([[0, 0], [2.75, 3], [2.75, 5], [2.75, 4.25], [1.5, 2.25], [2.75, 6], [-3.75, 2.75], [1.25, 4.5], [1.5, 3], [-3.25, 3], [-2.25, 0.25], [-0.25, 5.75], [0, 4.5]])

    greedy = greedy_traveling_salesman(points)    
    print("GREEDY")
    pprint(greedy)

    exhaustive = exhaustive_traveling_salesman(points)
    print("EXHAUSTIVE")
    pprint(exhaustive)


# GREEDY TRAVELING SALESMAN
# This function implements a simple heuristic: always go to the nearest node.
# This runs much more quickly and gives a solution only about 25% longer than the optimal one.
def greedy_traveling_salesman(points):
    num_stops = len(points)
    D = squareform(pdist(points))  #symmtric, square matrix of distances.  So D[i,j]=D[j,i]=distance from point i to point j
    stops_left = list(range(num_stops))
    greedy_path = []
    greedy_lengths = []

    current_stop = 0
    greedy_path.append(current_stop)
    stops_left.remove(current_stop)
    for _ in range(num_stops-1):
        idx = D[current_stop, stops_left].argmin()  #looks only at row for current_stop and only entries for stop_left.  Find the location of the shortest distance
        next_stop = stops_left[idx]  #idx is not the closest stop, it is the index of the closest stop.
        greedy_lengths.append(D[current_stop, next_stop])  #record length of this leg
        current_stop = next_stop
        greedy_path.append(current_stop)
        stops_left.remove(current_stop)
    greedy_length_total = sum(greedy_lengths)
    return {'path':greedy_path, 'lengths':greedy_lengths, 'length_total':greedy_length_total}


# EXHAUSTIVE TRAVELING SALESMAN
# This finds the true solution, but takes around 45 minutes to execute on CoCalc.
def exhaustive_traveling_salesman(points):
    num_stops = len(points)
    D = squareform(pdist(points))  #symmtric, square matrix of distances.  So D[i,j]=D[j,i]=distance from point i to point j
    best_path = []
    best_lengths = []
    best_length_total = D.sum()
    for stops in permutations(range(1,num_stops)):
        new_path = list(stops)
        new_path.insert(0, 0)
        legs = list(zip(new_path, new_path[1:]))
        new_lengths = [D[leg] for leg in legs]
        new_length_total = sum(new_lengths)
        if(new_length_total < best_length_total):
            best_path = new_path
            best_lengths = new_lengths
            best_length_total = new_length_total
    return {'path':best_path, 'lengths':best_lengths, 'length_total':best_length_total}

if __name__ == "__main__":
    main()
```

My next task was to create a program with the help of Dr. Cook that would test the exhaustive algorithm and greedy algorithm not just for the 13 buildings, but also for all possible subsets of those 13 buildings. In doing so, we can compare just how different the two algorithms may become when used on all 8,192 possible sets of buildings that a maintenance worker might have on their agenda to visit for a particular day. This code also eliminates the errors introduced by using the coordinate system, and instead calculates the travel time based on the initial recorded times rather than by Euclidean distance. The output for this code will provide three important pieces of information for each subset of buildings: the order in which to visit them the greedy way, the order in which to visit them the exhaustive way, and the percent time increase for the greedy path in comparison to the exhaustive path. The output orders these subsets by decreasing percent time difference.

<br></br>

```python
import doctest
import numpy as np
from scipy.spatial.distance import pdist, squareform
from pprint import pprint
import itertools as it
 
def greedy_traveling_salesman(Times, stops):
    greedy_path = []
    greedy_times = []
 
    current_stop = 0
    stops_left = stops.copy()
    try:
        stops_left.remove(current_stop)
    except:
        pass
    greedy_path.append(current_stop)
    num_stops = len(stops_left)
    for _ in range(num_stops):
        idx = Times[current_stop, stops_left].argmin()  #looks only at row for current_stop and only entries for stop_left.  Find the location of the shortest distance
        next_stop = stops_left[idx]  #idx is not the closest stop, it is the index of the closest stop.
        greedy_times.append(Times[current_stop, next_stop])  #record length of this leg
        current_stop = next_stop
        greedy_path.append(current_stop)
        stops_left.remove(current_stop)
    greedy_time_total = sum(greedy_times)
    return {'path':greedy_path, 'times':greedy_times, 'time_total':greedy_time_total}
 
def exhaustive_traveling_salesman(Times, stops):
    best_path = []
    best_times = []
    best_time_total = Times.sum()   
 
    current_stop = 0
    stops_left = stops.copy()
    try:
        stops_left.remove(current_stop)
    except:
        pass
    for stops in it.permutations(stops_left):
        new_path = list(stops)
        new_path.insert(0, current_stop)
        legs = list(zip(new_path, new_path[1:]))
        new_times = [Times[leg] for leg in legs]
        new_time_total = sum(new_times)
        if(new_time_total < best_time_total):
            best_path = new_path
            best_times = new_times
            best_time_total = new_time_total
    return {'path':best_path, 'times':best_times, 'time_total':best_time_total}
 
def main():  
    doctest.testmod()
   
    num_bldgs = 13
    #Times = np.random.rand(num_bldgs, num_bldgs)
    Times = [[0.00, 1.32, 2.32, 2.13, 1.19, 2.32, 4.52, 2.03, 1.40, 4.30, 1.07, 3.49, 4.01],
             [1.32, 0.00, 1.37, 1.13, 0.13, 1.37, 6.11, 0.51, 0.34, 6.02, 2.49, 3.16, 3.28],
             [2.32, 1.37, 0.00, 0.24, 1.28, 0.24, 8.21, 2.17, 1.14, 7.59, 6.18, 2.32, 2.44],
             [2.13, 1.13, 0.24, 0.00, 1.17, 1.07, 7.39, 1.24, 1.09, 7.14, 3.54, 3.10, 3.22],
             [1.19, 0.13, 1.28, 1.17, 0.00, 0.57, 6.04, 0.38, 0.21, 5.06, 2.04, 2.23, 2.35],
             [2.32, 1.37, 0.24, 1.07, 0.57, 0.00, 7.41, 3.05, 1.34, 7.59, 6.20, 2.51, 3.03],
             [4.52, 6.11, 8.21, 7.39, 6.04, 7.41, 0.00, 6.04, 6.25, 0.08, 3.45, 7.50, 8.02],
             [2.03, 0.51, 2.17, 1.24, 0.38, 3.05, 6.04, 0.00, 0.17, 5.33, 4.01, 1.46, 1.58],
             [1.40, 0.34, 1.14, 1.09, 0.21, 1.34, 6.25, 0.17, 0.00, 6.03, 2.04, 4.35, 4.47],
             [4.30, 6.02, 7.59, 7.14, 5.06, 7.59, 0.08, 5.33, 6.03, 0.00, 3.23, 7.28, 7.40],
             [1.07, 2.49, 6.18, 3.54, 2.04, 6.20, 3.45, 4.01, 2.04, 3.23, 0.00, 5.47, 5.59],
             [3.49, 3.16, 2.32, 3.10, 2.23, 2.51, 7.50, 1.46, 4.35, 7.28, 5.47, 0.00, 0.12],
             [4.01, 3.28, 2.44, 3.22, 2.35, 3.03, 8.02, 1.58, 4.47, 7.40, 5.59, 0.12, 0.00]]
    Times = np.array(Times)
    bldgs = list(range(1, num_bldgs))
    results = []
    for length in range(1, num_bldgs):
        for stops in it.combinations(bldgs, length):
            greedy = greedy_traveling_salesman(Times, list(stops))
            exhaustive = exhaustive_traveling_salesman(Times, list(stops))
            dif = greedy['time_total'] - exhaustive['time_total']
            percent_dif = dif / greedy['time_total'] * 100
            new_result = {'stops':stops, 'greedy':greedy, 'exhaustive':exhaustive, 'dif':dif, 'percent_dif':percent_dif}
            results.append(new_result)
    srt = np.argsort([res['percent_dif'] for res in results])
    for i in srt[::-1]:
        res = results[i]
        print("%d \t %.1f%% \t %s %s"%(i, res['percent_dif'], str(res['exhaustive']['path']).ljust(50), res['greedy']['path']))
 
if __name__ == "__main__":
    main()
```
